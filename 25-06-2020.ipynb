```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Boston Dataset
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston_df = pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
boston_df['MEDV'] = boston.target
boston_df.head()
```

```python
##to find the correlation we use corr()
boston_df.corr()
```

```python
#visualizing the correlation
sns.heatmap(boston_df.corr(),annot=True)
```

```python
plt.figure(figsize=(14,14))
sns.heatmap(boston_df.corr(),annot=True,cmap='YlGnBu')
```

```python
#Corelation>0.7
correlation=boston_df.corr()
cor=correlation[correlation>0.7]
plt.figure(figsize=(12,12))
sns.heatmap(cor,cmap='Greens',annot=True)
```

```python
#checking correlation for a given range
cor = correlation[(correlation>0.70)&(correlation<0.9)]
cor
```

```python
## All the columns--- as Input
```

```python
#split dat into training and testing;Bulid the model on traing data and test the model on testing data
```

```python
#split the data into Input and Output
x = boston_df.drop(['MEDV'],axis=1)
y = boston_df.MEDV
```

```python
#Divide the X and Y into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#Build the model onTraining data --->x_train and  y_train
#Sklearn library :import,instantiate,fit
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

### Checking the intercept and coefficient

```python
print(lm.intercept_)
lm.coef_
```

### creating a dataframe for coefficient values:

```python
coefficients = pd.DataFrame([x_train.columns, lm.coef_])
coefficients
```

```python
coefficients = pd.DataFrame([x_train.columns, lm.coef_]).T
coefficients
```

### Model Prediction training data:

```python
y_train_pred = lm.predict(x_train)
y_train_pred
```

```python
from sklearn.metrics import r2_score
print('R^2:', r2_score(y_train,y_train_pred))
```

```python
from sklearn.metrics import r2_score
print('R^2:', r2_score(y_train,y_train_pred))
print('Adjusted R^2:', 1- (1-r2_score(y_train, y_train_pred))*(len(x_train)-1)/
                                  (len(x_train)-x_train.shape[1]-1))
```

```python
## regplot(y_train-->actual values and y_train_pred-->predicted values)
sns.regplot(y_train,y_train_pred)
```

```python
## visualize the differences between actual and predicted values
plt.scatter(y_train,y_train_pred)
plt.xlabel('Price')
plt.ylabel('Model Predicted Price')
```

```python
plt.scatter(y_train_pred,y_train-y_train_pred)
plt.xlabel('Predicted values')
plt.ylabel('residual values')
```

```python
#testing the model on testing data
y_test_pred = lm.predict(x_test)#test data-->unseen data
y_test_pred
```

```python
#we need to compare the actual values(y_test) and the predicted values(y_test_pred)
from sklearn.metrics import r2_score ,mean_absolute_error,mean_squared_error
print('R^2',r2_score(y_test,y_test_pred))
print('Adjusted R^2:',1-(1-r2_score(y_test,y_test_pred))*(len(x_test)-1)/(len(x_test)-x_test.shape[1]-1))
print("MAE:",mean_absolute_error(y_test,y_test_pred))
print("MSE:",mean_squared_error(y_test,y_test_pred))
print('RMSE',np.sqrt(mean_squared_error(y_test,y_test_pred)))
```
